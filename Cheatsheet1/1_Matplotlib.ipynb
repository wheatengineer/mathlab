#### 1.1 Importing the library


```python
import matplotlib.pyplot as plt
```

#### 1.2 Setting up a graphic

Subplots creates a **figure** and **axes**, while the *axes* are used to plot with x/y coordinates later, the *figure* is used to display the entire graphic as configured with all the plots previously made using the "display()" function.

The sizes here are used for the actual size of the graphic and *not* for the maximum range of the plottable points.

```python
size_x, size_y = 8, 6

fig, ax = plt.subplots(figsize=(size_x, size_y))
```

#### 1.3 Configuring the Graphic

**1.3.1 Equalizing the aspect ratio**

The function "ax.set_aspect" is used to equalize the aspect ratio of the x- and y-values to ensure correct graphical representation.  


```python
ax.set_aspect('equal')

display(fig)
```

**1.3.2 Axis range and ticks**  

"ax.axis()" is used with an array input to configure the maximum range of each of the defined axes.

"ax.set_xticks()" and "ax.set_yticks()" is used with an array input to configure which ticks are visible on the respective axis.  
(For this we will be using "np.arange()", see the Numpy cheatsheet)

"ax.set_xticks()" and "ax.set_yticks()" have a modifier after their initial array input which is "minor=True" (with no modifier it is "major"). This makes 

```python
import numpy as np

x_min, x_max, y_min, y_max = -15, 15, -10, 10

ax.axis([x_min, x_max, y_min, y_max])

x_ticks = np.arange(x_min, x_max+1, 3)
y_ticks = np.arange(y_min, y_max+1, 2)
ax.set_xticks(x_ticks)
ax.set_yticks(y_ticks)

minor_x_ticks = np.arange(x_min, x_max+1, 1.5)
minor_y_ticks = np.arange(y_min, y_max+1, 1)
ax.set_xticks(minor_x_ticks, minor=True)
ax.set_yticks(minor_y_ticks, minor=True)

display(fig)
```
